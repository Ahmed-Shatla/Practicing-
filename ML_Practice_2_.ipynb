```python
car_sales = pd.read_csv("/content/car-sales-extended.csv")
car_sales
```

```python
car_sales.dtypes
```

```python
#Make sure all data is numircal
X = car_sales.drop(['Price'],axis=1)
y = car_sales['Price']
```

```python
from math import remainder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_col = ["Make","Colour","Doors"]
OneHot = OneHotEncoder()
transformer = ColumnTransformer( [("One_Hot",
                                  OneHot,
                                  categorical_col)],
                                  remainder="passthrough")
transformed_X = transformer.fit_transform(X)
```

```python
df = pd.DataFrame(transformed_X)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
car_missing = pd.read_csv("/content/car-sales-extended-missing-data.csv")
car_missing
```

```python
car_missing.isna().sum()
```

```python
car_missing.dropna(subset=["Price"],inplace=True)
car_missing.isna().sum()
```

```python
X = car_missing.drop(['Price'],axis=1)
y = car_missing['Price']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#fill catigorical with "missing" and numerical with mean
ctg_imputer = SimpleImputer(strategy="constant",fill_value="missing")
door_imputer = SimpleImputer(strategy="constant",fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

ctg_col = ['Make','Colour']
door_col = ['Doors']
num_feature = ["Odometer (KM)"]

imputer = ColumnTransformer([("ctg",ctg_imputer,ctg_col),
                             ("door",door_imputer,door_col),
                             ("cum",num_imputer,num_feature)])

filled_X = imputer.fit_transform(X)
filled_X
```
